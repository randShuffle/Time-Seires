```python
import os

# 设置您想要列出文件的目录的路径
folder_path = '../diabetes_datasets/Shanghai_T2DM_column_unified_processed/Shanghai_T2DM_column_unified_processed/'

# 获取文件夹下的所有文件和文件夹名称
files_and_folders = os.listdir(folder_path)

# 如果您只想列出文件，可以进一步过滤
files = [f for f in files_and_folders if os.path.isfile(os.path.join(folder_path, f))]

# files.remove('.DS_Store')
# 打印所有文件名
print(files)
```

```python
import pandas as pd

# hour_sin
# hour_cos
# CGM_diff

output_path = '../diabetes_datasets/T2'
def preprocess(filename:str):
    df=pd.read_excel(folder_path+'/'+filename)
    df['insulin'] = df[['Insulin dose - s.c.', 'Non-insulin hypoglycemic agents', 'CSII - bolus insulin (Novolin R, IU)', 'CSII - basal insulin (Novolin R, IU / H)', 'Insulin dose - i.v.']].notnull().any(axis=1).astype(int)
    df['insulin']=df['insulin'].fillna(0)
    df['carbohydrate']=df['carbohydrate'].fillna(0)
    df['protein']=df['protein'].fillna(0)
    df['fat']=df['fat'].fillna(0)
    df['cellulose']=df['cellulose'].fillna(0)
    df=df[['CGM (mg / dl)','insulin','carbohydrate','protein','fat','cellulose','hour_sin','hour_cos','CGM_diff']]
    df.dropna(subset=['CGM_diff'], inplace=True)
    name, ext = os.path.splitext(output_path+'/'+filename)
    new_filename = name + '.csv'
    df.to_csv(new_filename,index=False)

for file in files:
    preprocess(file)
    print(file)
   
```
